```python
from scipy.io import loadmat
from scipy.ndimage import uniform_filter1d
import matplotlib.pyplot as plt


# Load the .mat file to check its structure
mat = loadmat("../dataset/AR1a_W.mat")

# Print all the keys in the loaded .mat file to understand its structure
mat_keys = list(mat.keys())
mat_keys
```

```python
# Extract the 'csi_buff' data from the .mat file
csi_data = mat["csi_buff"]

# Display basic information about the CSI data to understand its structure
csi_data_info = {
    "type": type(csi_data),
    "shape": csi_data.shape,
}

csi_data_info
```

```python
# Apply a moving average filter with a window size of 5
window_size = 5
filtered_csi_data = uniform_filter1d(csi_data, size=window_size, axis=0)

# Compare a sample of the original and filtered data to see the effect
sample_index = 200  # Arbitrary sample index for illustration

original_sample = csi_data[sample_index, :]
filtered_sample = filtered_csi_data[sample_index, :]

# Plotting the original and filtered data for comparison
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(original_sample)
plt.title("Original CSI Data Sample")
plt.xlabel("Feature Index")
plt.ylabel("Value")

plt.subplot(1, 2, 2)
plt.plot(filtered_sample)
plt.title("Filtered CSI Data Sample")
plt.xlabel("Feature Index")
plt.ylabel("Value")

plt.tight_layout()
plt.show()
```
